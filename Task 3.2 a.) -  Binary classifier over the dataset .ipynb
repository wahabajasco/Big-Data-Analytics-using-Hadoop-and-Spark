```python
#Replace null values in attack_cat column with normal
dataset = UNSWNB15.na.fill(value='normal',subset=["attack_cat"])
dataset.show()
```

```python
#Update all values in attack_cat column
#from pyspark.sql.functions import col,when
from pyspark.sql.functions import col,when
updated_dataset = dataset.withColumn("attack_cat", when( col("attack_cat")=="normal","Normal" )\
.when( col("attack_cat")=="Generic","Generic" )\
.when( col("attack_cat")=="Exploits","Exploits" )\
.when( col("attack_cat")=="DoS","DoS" )\
.when( col("attack_cat")=="Fuzzers","Fuzzers" )\
.when( col("attack_cat")=="Analysis","Analysis" )\
.when( col("attack_cat")=="Reconnaissance","Reconnaissance" )\
.when( col("attack_cat")=="Backdoors","Backdoors" )\
.when( col("attack_cat")=="Shellcode","Shellcode" )\
.when( col("attack_cat")=="Worms","Worms" )\
.when( col("attack_cat")=="Fuzzers ", "Fuzzers")\
.when( col("attack_cat")=="Reconnaissance ","Reconnaissance" )\
.when( col("attack_cat")=="Backdoor","Backdoors" )\
.when( col("attack_cat")=="Shellcode ","Shellcode" )\
)
```

```python
#Confirm the dataset and the number of records
updated_dataset.show()
updated_dataset.count()
```

```python
#Randomly split data into training and test sets, and set seed for reproducibility.
trainDF, testDF = dataset.randomSplit([0.8, 0.2], seed=42)
print(trainDF.cache().count()) # Cache because accessing training data multiple times
print(testDF.count())
```

```python
#Import StringIndexer and OneHotEncoder
from pyspark.ml.feature import StringIndexer, OneHotEncoder

categoricalCols = ["srcip", "dstip", "proto", "state"]
# The following two lines are estimators. They return functions that we will later apply to transform the dataset.
stringIndexer = StringIndexer(inputCols=categoricalCols, outputCols=[x + "Index" for x in categoricalCols]) 
encoder = OneHotEncoder(inputCols=stringIndexer.getOutputCols(), outputCols=[x + "OHE" for x in categoricalCols]) 
# The label column ("attack_cat") is also a string value 
# Convert it to a numeric value using StringIndexer.
labelToIndex = StringIndexer(inputCol="attack_cat", outputCol="label")
```

```python
stringIndexerModel = stringIndexer.fit(trainDF)
display(stringIndexerModel.transform(trainDF))
```

```python
#Import VectorAssembler
from pyspark.ml.feature import VectorAssembler
# This includes both the numeric columns and the one-hot encoded binary vector columns in the dataset.
numericCols = ["sport", "dsport", "trans_depth", "dur","Spkts", "Dpkts"]
assemblerInputs = [c + "OHE" for c in categoricalCols] + numericCols
vecAssembler = VectorAssembler(inputCols=assemblerInputs, outputCol="features")
```

```python
#Import LogisticRegression
from pyspark.ml.classification import LogisticRegression
#Model defination using LogisticRegression
lr = LogisticRegression(featuresCol="features", labelCol="label", regParam=1.0)
```

```python
#Import Pipeline
from pyspark.ml import Pipeline
# Define the pipeline based on the stages created in previous steps.
pipeline = Pipeline(stages=[stringIndexer, encoder, labelToIndex, vecAssembler, lr])
# Define the pipeline model.
pipelineModel = pipeline.fit(trainDF)
# Apply the pipeline model to the test dataset.
predDF = pipelineModel.transform(testDF)
```

```python
#Display the predictions from the model. 
display(predDF.select("features", "label", "prediction", "probability"))
predDF.select("features", "label", "prediction", "probability").show()
```

```python
display(pipelineModel.stages[-1], predDF.drop("prediction", "rawPrediction", "probability"), "ROC")
```

```python
#Import MulticlassClassificationEvaluator
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
#Using the MulticlassClassificationEvaluator to verify the accuracy.
mcEvaluator = MulticlassClassificationEvaluator(metricName="accuracy")
print(f"Accuracy:{mcEvaluator.evaluate(predDF)}")
```

```python
#Import BinaryClassificationEvaluator
from pyspark.ml.evaluation import BinaryClassificationEvaluator
#Using the BinaryClassificationEvaluator to evaluate the area under the ROC curve
bcEvaluator = BinaryClassificationEvaluator(metricName="areaUnderROC")
print(f"Area under ROC curve:{bcEvaluator.evaluate(predDF)}")
```
